```python
a = 3
b = 5
c = 7
result = not a > b and b < c or c == 7
print(result)
```

True and True  # 结果为 True
True and False  # 结果为 False
False and False  # 结果为 False
True or False  # 结果为 True
False or False  # 结果为 False
not True  # 结果为 False
not False  # 结果为 True

首先计算not运算符（优先级最高）：
    not a > b 等价于 not (a > b)，首先计算 a > b：
        a 是 3，b 是 5，a > b 结果为 False。
        然后 not False，结果为 True。
接着计算 and 运算符（优先级次高）：
    现在我们计算 True and b < c，其中 b < c 等价于 5 < 7，结果为 True。
    因此，True and True 结果为 True。
最后计算 or 运算符（优先级最低）：
    最终的表达式是 True or c == 7，其中 c == 7 为 True。
    因为 or 只要有一个为 True，最终结果就是 True。
